## Affichage des donn√©es
Afin de visualiser correctement le contenu des informations pr√©sentent dans la partie `stations` de la structure JSON, vous devez r√©aliser une mise en forme des donn√©es afin qu'elles soient affich√©es sous forme d'un tableau dont les ent√™tes de colonnes indiquent le nom des champs (cl√©s).

üêº Pour cela, on utilise la m√©thode `DataFrame` de la biblioth√®que Pandas :

```python
# afficher les donn√©es sous forme d'un Data Frame
import pandas as pd

stations_ = pd.DataFrame(json_data['data']['stations'])

display(stations_)
```

## Analyse de valeurs

Comme on peut le voir √† l'affichage pr√©c√©dent, chaque station est constitu√©e de diff√©rentes rubriques : `station_id`,	`name`,	`lat`,	`lon`,	`capacity`,	`stationCode` et `rental_methods`.

‚ö† Ici, nous nous int√©resserons √† la colonne `capacity` qui indique le nombre de stands par station pouvant acceuillir les v√©los.

Ce code permet de r√©cup√©rer et d'afficher les donn√©es de la colonne `capacity`.

```python
capacities =pd.DataFrame(stations_['capacity'])
display(capacities)
```

La variable `capacities` est un tableau de valeurs num√©riques qui repr√©sentent la capacit√© de chaque station V√©lib.

On demande de produire des mesures qui indiquent :
1. le nombre total de stations de stands √† v√©los : m√©thode `count()` de la variable `capacities`,
2. le nombre total de stands (libres ou occup√©s) √† v√©los sur l'ensemble des stations : m√©thode `sum()` de la variable `capacities`,
3. le nombre minimum de stands : m√©thode `min()` de la variable `capacities`,
4. le nombre maximum de stands : m√©thode `max()` de la variable `capacities`,
5. le nombre moyen de stands : m√©thode `mean()` de la variable `capacities`,
6. l'√©cart-type du nombre de stands : m√©thode `std()` de la variable `capacities`,

Chacune de ces mesures devra √™tre m√©moris√©e dans une variable num√©rique.

```python
# analyser les valeurs

# nombre de stations
nb_stations = int(capacities.count())
# nombre de stands
nb_stands = int(capacities.sum())
# nombre minimum de stands
nb_min_stands = int(capacities.min())
# nombre maximum de stands
nb_max_stands = int(capacities.max())
# nombre moyen de stands
nb_moyen_stands = float(capacities.mean())
# √©cart-type de stands
nb_ecarttype_stands = float(capacities.std())
```

## Publication des r√©sultats

Finalement, proposez une publication des mesures obtenues dans un format texte lisible et compr√©hensible. L'affichage se fait dans la console.

```python
# publier les r√©sultats
print('nombre de stations :', nb_stations )
print('nombre de stand :', nb_stands )
print('nombre minimum de stand :', nb_min_stands )
print('nombre maximum de stand :', nb_max_stands )
print('nombre moyenne de stand :', nb_moyen_stands )
print('nombre moyenne de stand :', nb_ecarttype_stands )
```
