## Lecture des donn√©es V√©lib

```python
# chargement des donn√©es de status et d'information depuis l'API V√©lib' M√©tropole (fonctions supports)
info = tools.loadVelibInformation()
differentes_info = tools.getVelibStations(info)


infos = tools.loadVelibStatus()
differentes_infos = tools.getVelibStations(infos)
```

## Pr√©paration des donn√©es

```python
# conversion en Data Frames Pandas des donn√©es de status et d'information (pd.DataFrame)
affiche = pd.DataFrame(differentes_info)


affichage = pd.DataFrame(differentes_infos)

# fusion des deux Data Frames (fonction merge)
a=affiche.merge(affichage)

# affichage du r√©sultat
display(a)
```

## Analyses statistiques

```python
# mesures statistiques assembl√©es (fonctions describe et join)
param = pd.DataFrame(a.station_id.describe())
param = param.join(a.capacity.describe())
param = param.join(a.numBikesAvailable.describe())
param = param.join(a.numDocksAvailable.describe())
display(param)
```

## Exportation pour le Web

```python
# renommage des colonnes avant l'exportation en HTML (fonction rename)

b=param.rename(columns={
    "capacity" : "nombres bornettes dans la station",
    "numBikesAvailable" : "nombre de v√©los dispo",
    "numDocksAvailable" : "nombre de stands disponibles"})


# exportation des mesures au format HTML (fonctions open et to_html)

f=open("/content/drive/My Drive/Colab Notebooks/SAE15/web/data/stati.html","w") #cette ligne ne modifie pas un fichier mais cree un nouveau fichier contenant notre html
f.write(b.to_html())


```
